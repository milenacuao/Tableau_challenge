```python
# importing Dependencies 
import pandas as pd
import numpy as np 
```

```python
# uncleaned CSV files
csv1 = "/Users/rociocuao/Desktop/2019-citibike-tripdata/11_November/201911-citibike-tripdata_2.csv"
csv2 ="/Users/rociocuao/Desktop/2020-citibike-tripdata/11_November/202011-citibike-tripdata_2.csv"
citi_bike_trip_2019 = pd.read_csv(csv1)
citi_bike_trip_2020 = pd.read_csv(csv2)
```

```python
# 2019 Data
citi_2019_df = citi_bike_trip_2019
citi_2019_df.head(2)
```

```python
# 2020 Data
citi_2020_df = citi_bike_trip_2020
citi_2020_df.head(2)
```

```python
# Deleting Columns
del citi_2019_df['usertype']
del citi_2020_df['usertype']
```

```python
citi_2019_df.head(6)
citi_2020_df.head(6)
```

```python
# Checking Type of Values
citi_2019_df.dtypes
citi_2020_df.dtypes
```

```python
# Changing The Type of Values for the Gender Column
citi_2019_df['gender'] = citi_2019_df['gender'].astype(str)
citi_2020_df['gender'] = citi_2020_df['gender'].astype(str)
```

```python
# Changing the Numeric Gender Values to Actual Gender Values For The 2018 CSV
gender_check_2019 = citi_2019_df['gender']
changed_2019_male = gender_check_2019.replace('1', 'Male')
changed_2019_female = changed_2019_male.replace('2', 'Female')
changed_2019_unknown = changed_2019_female.replace('0', 'Unknown')
citi_2019_df['gender'] = changed_2019_unknown
citi_2019_df
```

```python
# Changing the Numeric Gender Values 2020 
gender_check_2020 = citi_2020_df['gender']
changed_2020_male = gender_check_2020.replace('1', 'Male')
changed_2020_female = changed_2020_male.replace('2', 'Female')
changed_2020_unknown = changed_2020_female.replace('0', 'Unknown')
citi_2020_df['gender'] = changed_2020_unknown
citi_2020_df
```

```python
#  Age's of Bikers 
current_year = 2019

birth_year_2019 = citi_2019_df['birth year']
age = current_year - birth_year_2019
citi_2019_df['age'] = age

birth_year_2020 = citi_2020_df['birth year']
age = current_year - birth_year_2020
citi_2020_df['age'] = age
```

```python
# Rearranging the Columns
citi_2019_df_organized = citi_2019_df[["bikeid",
                                       "gender",
                                       "birth year",
                                       "age",
                                       "starttime", 
                                       "stoptime",
                                       "tripduration",
                                       "start station id",
                                       "end station id",
                                       "start station name",
                                       "end station name",
                                       "start station latitude",
                                       "end station latitude",
                                       "start station longitude",
                                       "end station longitude"]]

citi_2020_df_organized = citi_2020_df[["bikeid",
                                       "gender",
                                       "birth year",
                                       "age",
                                       "starttime", 
                                       "stoptime",
                                       "tripduration",
                                       "start station id",
                                       "end station id",
                                       "start station name",
                                       "end station name",
                                       "start station latitude",
                                       "end station latitude",
                                       "start station longitude",
                                       "end station longitude"]]

citi_2019_df_organized.head(10)
citi_2020_df_organized.head(10)
```

```python
# Renaming the Columns
citi_2019_renamed_df = citi_2019_df_organized.rename(columns={"bikeid":"Bike ID", 
                                                              "gender":"Gender",
                                                              "birth year":"Birth Year",
                                                              "age":"Age",
                                                              "tripduration":"Trip Duration",
                                                              "starttime":"Start Time",
                                                              "stoptime":"Stop Time",
                                                              "start station id":"Start Station ID",
                                                              "end station id":"End Station ID",
                                                              "start station name":"Start Station Name",
                                                              "end station name":"End Station Name",
                                                              "start station latitude":"Start Station Latitude",
                                                              "end station latitude":"End Station Latitude",
                                                              "start station longitude":"Start Station Longitude",
                                                              "end station longitude":"End Station Longitude"})

citi_2020_renamed_df = citi_2020_df_organized.rename(columns={"bikeid":"Bike ID", 
                                                              "gender":"Gender",
                                                              "birth year":"Birth Year",
                                                              "age":"Age",
                                                              "tripduration":"Trip Duration",
                                                              "starttime":"Start Time",
                                                              "stoptime":"Stop Time",
                                                              "start station id":"Start Station ID",
                                                              "end station id":"End Station ID",
                                                              "start station name":"Start Station Name",
                                                              "end station name":"End Station Name",
                                                              "start station latitude":"Start Station Latitude",
                                                              "end station latitude":"End Station Latitude",
                                                              "start station longitude":"Start Station Longitude",
                                                              "end station longitude":"End Station Longitude"})



citi_2019_renamed_df
citi_2020_renamed_df
```

```python
citi_2019_renamed_df.to_csv('citibike_19.csv')
citi_2020_renamed_df.to_csv('citibike_20.csv')
```
